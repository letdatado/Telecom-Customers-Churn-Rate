```python
# Import libraries
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Column and row view settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

```

```python
# Read in the dataset
DATA_PATH = r"C:\Users\ALI\Projects\Telco Customers Churn\data\raw\Telco_customer_churn.xlsx"
df = pd.read_excel(DATA_PATH)
display(df.head())
```

```python
# General information about the dataset
info_df = pd.DataFrame({
    "column": df.columns,
    "non_null_count": df.notnull().sum().values,
    "dtype": df.dtypes.values
})

display(info_df)
```

```python
display(df.describe(include="all").T)
```

```python
df.columns
```

```python
contract_churn = (
    df.groupby("Contract")["Churn Label"]
    .value_counts(normalize=True)
    .rename("rate")
    .reset_index()
)

px.bar(
    contract_churn,
    x="Contract",
    y="rate",
    color="Churn Label",
    barmode="group",
    title="Churn Rate by Contract Type",
    text_auto=".1%"
)
```

```python
px.histogram(
    df,
    x="Tenure Months",
    color="Churn Label",
    nbins=40,
    barmode="overlay",
    title="Tenure Distribution by Churn Status"
)
```

```python
px.box(
    df,
    x="Churn Label",
    y="Monthly Charges",
    title="Monthly Charges by Churn Status"
)
```
